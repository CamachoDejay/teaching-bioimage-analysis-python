```python
import napari
from napari.utils import nbscreenshot
```

Creating a viewer:

```python
viewer = napari.Viewer()
# as the viewer is an independent ui, we use the screenshot function to store
# an image of the viewer in our notebook
nbscreenshot(viewer)
```

skimage is one of the msot useful tools around for image processing
https://scikit-image.org/

```python
import skimage.io as skio
```

Now lets read an image an control the layers of napari

```python
im = skio.imread("data/rice.tif", plugin="tifffile")
viewer.add_image(im, name = 'rice')
viewer.layers['rice'].colormap = 'gray'
nbscreenshot(viewer)
```

Using the viewer methods we can then automate the loading and display of large data. As a community we are also working on advanced image formats as the next generation file format https://ngff.openmicroscopy.org/0.1/

```python
lbl = skio.imread("data/rice_lbl.tif", plugin="tifffile")
# notice how the properties of a layer correspond with the information in the ui
viewer.add_labels(lbl, name = 'rice_label', blending='translucent')
nbscreenshot(viewer)
```
