# Book 2: How to do simple plots with Pandas

Here we will go over how to use ```pandas``` to read ```.csv``` files and then make a simple plot

## Load the data
Let us use what we learned before to load the ```.csv``` table into a data frame

````python
import pandas as pd 

df1 = pd.read_csv('./data/Results_01.csv')

df1.head()
````

## Using Pandas to make basic plots

Pandas can be used to do many statistical operations but also to do basic plots easily. To do this we will need to install another well known package called [matplotlib](https://matplotlib.org/). As an exercise you can try to run the cell below before and after intalling ```matplotlib```

````python
# simple histogram
ax = df1['Area'].plot.hist(bins=20, edgecolor='black')
ax.set_ylabel('Occurrence')
ax.set_xlabel('Area')
ax.set_title('Histogram of Area')
````

Here we have a bit more control over the plot, and we init the figure directly from matplotlib

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df1['Area'], bins=20, edgecolor='black')
plt.ylabel('Occurrence')
plt.xlabel('Area')
plt.title('Histogram of Area')
````

Now, we'll use a different plotting library called Seaborn. Seaborn is a lighter-weight alternative to Matplotlib, offering an easy-to-use interface and beautiful default styles. It is particularly well-suited for statistical visualizations and comes with a variety of built-in themes. For inspiration and examples, you can explore its comprehensive [gallery here](https://seaborn.pydata.org/examples/index.html).

````python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df1, x='Area', kde=True, bins=20)
plt.ylabel('Occurrence')
plt.title('Histogram of Area')
````
