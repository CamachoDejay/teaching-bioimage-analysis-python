```python
import napari
from napari.utils import nbscreenshot
```

Creating a viewer:

```python
viewer = napari.Viewer()
nbscreenshot(viewer)
```

skimage is one of the msot useful tools around for iamge processing
https://scikit-image.org/

```python
import skimage.io as skio
```

Now lets read an image an control the layers fo napari

```python
im = skio.imread("data/telomeres/C1-condition_1.tif", plugin="tifffile")
viewer.add_image(im, name = 'telo')
viewer.layers['telo'].colormap = 'magenta'
nbscreenshot(viewer)
```

By knowing a few things then we can automated the loading and display of large data, we are also working on advanced image formats as the next generation file format https://ngff.openmicroscopy.org/0.1/

```python
im1 = skio.imread("data/telomeres/C1-condition_1.tif", plugin="tifffile")
im2 = skio.imread("data/telomeres/C2-condition_1.tif", plugin="tifffile")
viewer.add_image(im1, name = 'C1',colormap='magenta',blending='additive')
viewer.add_image(im2, name = 'C2',colormap='green', blending='additive')
nbscreenshot(viewer)
```
